```python
seed(42)
print(uniform(0, 1))
```

```python
csucsok = [[0, 0], [2, 0], [1, 1.73]]
points = [csucsok[0]]

def felezo(t1, t2):
    elso = (t1[0] + t2[0])/2
    masodik = (t1[1] + t2[1])/2
    return [elso, masodik]



for _ in range(50000):
    last_point = points[-1]
    new_point = choice(csucsok)
    points.append(felezo(last_point, new_point))
    
coordinates = [[], []]
for coordinate in points:
    coordinates[0].append(coordinate[0])
    coordinates[1].append(coordinate[1])
plt.scatter(coordinates[0], coordinates[1])
plt.show()
```
