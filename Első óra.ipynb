# Sorting

## Bubble sort

```python
n = 10000
A = [i for i in range(n)]
shuffle(A)
start = time.time()
for i in range(n):
    for j in range(i, n):
        if A[i] > A[j]:
            A[i], A[j] = A[j], A[i]
end = time.time()
print(end - start)
```

## Merge sort

```python
n = 10
A = [i for i in range(n)]
shuffle(A)

def merge(l1, l2):
    merged = []
    i, j = 0, 0
    while i < len(l1) and j < len(l2):
        if l1[i] < l2[j]:
            merged.append(l1[i])
            i += 1
        else:
            merged.append(l2[j])
            j += 1
    if i < len(l1):
        merged.extend(l1[i:])
    elif j < len(l2):
        merged.extend(l2[j:])
        
    return merged
            
def merge_sort(list_, l, r):
    if l < r:
        m = l + (r - l) // 2
    
        merge_sort(list_, l, m)
        merge_sort(list_, m+1, r)
        merge(list_[l:m], list_[m+1,r])

print(merge_sort(A, 0, n-1))
```
